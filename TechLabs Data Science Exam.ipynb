E) Briefly explain what a lambda function is and its benefits over using regular functions.

#### E Answer)
A Lambda function (also called anonymous function) is not bound to an identifier. It is oftentimes used to pass to higher-order functions (those that return a function) and in cases where its purpose is single-use. This leads to greater readability of the code as the top-down flow is not interrupted. Using ordinary functions for single-use cases interrups the flow and complicated readability. Especially if such single-use cases are occuring often.


F) What is the difference between relational and non-relational databases?

#### F Answer)
Relational databases link data that is stored in tables and rows through so-called "keys". This ensures accuracy and integrity of data, also called "referential integrity. Use-cases include Structured Querying Language (SQL) which can be used to update, edit, and retrieve records from a relational database.
Unrelated databases also stores data, but not in table and rows. Instead it uses one of four models each optimized for specific requirements. Use-cases include for example highly related data and "Big Data" (= large amounts of unstructured data). These systems are highly flexible and allow for future scaling.
